# Cylindrical Coordinates

```python
def xyz_to_cylindrical(x, y, z):
    """
    Convierte coordenadas cartesianas (x, y, z) a cilÃ­ndricas (phi, rho, z).
    
    ParÃ¡metros:
        x, y, z : float o array_like
            Coordenadas cartesianas.
    
    Retorna:
        phi : float o np.ndarray
            Ãngulo azimutal en radianes, entre -pi y pi.
        rho : float o np.ndarray
            Distancia radial al eje z.
        z : float o np.ndarray
            Misma coordenada z de entrada.
    """
    rho = np.sqrt(x**2 + y**2)
    phi = np.arctan2(y, x)
    
    return phi, rho, z
```
